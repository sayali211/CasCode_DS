```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
student_data=pd.read_csv(r'C:\Users\Sushant\Downloads\archive\student_data.csv')
```

```python
print(student_data.head())
```

```python
summary_status=student_data.describe()
print(summary_status)
```

```python
data_info=student_data.info()
print(data_info)
```

```python
unique_values=student_data.select_dtypes(include='object').nunique()
print(unique_values)
```

```python
correlation_matrix=student_data.corr()
```

```python
print(correlation_matrix)
```

```python
unique_counts=student_data.select_dtypes(include='number').nunique()
print(unique_counts)
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(student_data['age'], bins=20, kde=True)
plt.title('Distribution of age')
plt.xlabel('age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(x='Mjob',y='age',data=student_data)
plt.title("Mjob vs Age")
plt.xlabel("Mjob")
plt.ylabel("Age")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='age',y='Mjob',data=student_data, hue="failures")
plt.title('Scatter Plot of Age vs Mjob')
plt.xlabel('Age')
plt.ylabel('Mjob')
plt.legend(title='Gender')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='studytime', data=student_data)
plt.title('Study time Distribution')
plt.xlabel('Study time')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```
